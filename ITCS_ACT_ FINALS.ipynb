# 5. CAN WE IDENTIFY ANY SEASONAL PATTERNS IN THE EMPLOYMENT AND UNEMPLOYMENT RATES?

```python
#visualization using scatterplot

employdata['Month'] = pd.Categorical(employdata['Month'], categories=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November','December'], ordered=True)
df_sorted = employdata.sort_values('MonthNumber')

sns.scatterplot(data=df_sorted, x= 'Employment', y= 'Month', label='Employment')
sns.scatterplot(data=df_sorted, x= 'Unemployment', y='Month', label='Unemployment')
plt.title('Seasonal Patterns in Employment and Unemployment Rates')
plt.xlabel('Rate')
plt.ylabel('Month')
plt.legend()
plt.gcf().set_size_inches(8,10)
plt.show()
```

    <br>
    CONCLUSION:
    <br>
         According to Blau, D. M., & Robins, P. K. (1988), Seasonal patterns can manifest as month-to-month variations in employment and unemployment rates. You might observe consistent increases or decreases in rates during particular months across multiple years. This can be influenced by factors such as holidays, school calendars, or weather conditions.
