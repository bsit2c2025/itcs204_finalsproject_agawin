# 1. WHAT DATE HAS THE HIGHEST AMOUNT OF ORDER?

```python
# Count orders by date and find the date with highest orders
orders_by_date = amazondata.groupby('OrderDate')['OrderID'].count().reset_index()
orders_by_date.columns = ['OrderDate', 'OrderCount']
orders_by_date = orders_by_date.sort_values('OrderCount', ascending=False)

# Plot
sns.lineplot(data=orders_by_date, x='OrderDate', y='OrderCount', color='green', marker='o')
plt.title('Date With the Highest Number of Orders')
plt.xlabel('Order Date')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the date with highest orders
print(f"Date with highest orders: {orders_by_date.iloc[0]['OrderDate']} with {orders_by_date.iloc[0]['OrderCount']} orders")
```

**INSIGHTS** : According to the Visualization above, this shows an Highest number of Orders with 85 orders that happened on April 28, 2022 but the order date is not properly showned because it's dates are stacked by one another. But using code that extracts the value of the date with the highest order, it shows that 85 orders happened on April 28, 2022.

# 2. HIGHEST FREQUENCY OF A CUSTOMER NAME 

```python
# Compute customer order frequency and visualize the top customers
customer_counts = amazondata['CustomerName'].value_counts().reset_index()
customer_counts.columns = ['CustomerName', 'OrderCount']

# Top 10 customers by order count
top_customers = customer_counts.head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_customers, x='OrderCount', y='CustomerName', palette='viridis')
plt.title('Top 10 Customers by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('Customer Name')
for i, v in enumerate(top_customers['OrderCount']):
    plt.text(v + 0.5, i, str(v), va='center')
plt.tight_layout()
plt.show()

# Print most frequent customer(s)
max_count = customer_counts['OrderCount'].max()
most_frequent = customer_counts[customer_counts['OrderCount'] == max_count]
print("Most frequent customer(s):")
print(most_frequent)
```

**INSIGHTS** : Visualization using a bar plot that shows the most frequent customer the data of the Amazon has. Having Karan Joshi with the order count of 556. Our top 5 costumers is Karan Joshi, Arjun Kumar, Pooja Kapoor, Rohit Gupta, and Vihaan Singh.

# 3. What month/year generates the highest total revenue?

```python
# Convert OrderDate to datetime and aggregate revenue by month-year
amazondata['OrderDate'] = pd.to_datetime(amazondata['OrderDate'])
amazondata['MonthYear'] = amazondata['OrderDate'].dt.to_period('M').dt.to_timestamp()
revenue_by_month = amazondata.groupby('MonthYear')['TotalAmount'].sum().reset_index().sort_values('MonthYear')

# Identify top month-year
top = revenue_by_month.loc[revenue_by_month['TotalAmount'].idxmax()]

# Plot revenue by month-year and highlight the highest
plt.figure(figsize=(14,6))
colors = ['orange' if dt == top['MonthYear'] else 'skyblue' for dt in revenue_by_month['MonthYear']]
plt.bar(revenue_by_month['MonthYear'].dt.strftime('%Y-%m'), revenue_by_month['TotalAmount'], color=colors)
plt.xticks(rotation=45)
plt.ylabel('Total Revenue (TotalAmount)')
plt.title('Monthly Revenue by Month-Year')
plt.tight_layout()
plt.show()

print(f"Month/year with highest total revenue: {top['MonthYear'].strftime('%Y-%m')} with ${top['TotalAmount']:.2f}")
```

**INSIGHTS** : The month of May 2024 (2024-05) recorded the highest total revenue, amounting to $1,642,609.94. This indicates a strong surge in customer purchasing activity during this period, which may be attributed to factors such as seasonal demand, promotional campaigns, or increased product availability. The peak in revenue suggests that May is a critical month for sales performance, and Amazon could benefit from reinforcing marketing efforts, inventory readiness, and logistics planning during this period to maximize future revenue potential.

# 4. Are there seasonal trends in order volume?

```python
# Seasonal trends in order volume: heatmap (month vs year) and average monthly orders across years
amazondata['OrderYear'] = amazondata['OrderDate'].dt.year
amazondata['OrderMonth'] = amazondata['OrderDate'].dt.month

monthly_counts = amazondata.groupby(['OrderYear', 'OrderMonth'])['OrderID'].count().reset_index(name='OrderCount')
pivot = monthly_counts.pivot(index='OrderMonth', columns='OrderYear', values='OrderCount').reindex(range(1,13))

# Heatmap + Avg monthly orders
plt.figure(figsize=(16,6))

# Heatmap
plt.subplot(1,2,1)
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlGnBu', cbar_kws={'label':'Order Count'}, linewidths=.5)
plt.title('Monthly Order Volume (Month vs Year)')
plt.xlabel('Year')
plt.ylabel('Month')
month_labels = pd.to_datetime(pivot.index.astype(str), format='%m').month_name()
plt.yticks(np.arange(pivot.shape[0]) + 0.5, labels=month_labels, rotation=0)

# Average monthly orders with variability
plt.subplot(1,2,2)
avg_orders = monthly_counts.groupby('OrderMonth')['OrderCount'].mean().reindex(range(1,13))
std_orders = monthly_counts.groupby('OrderMonth')['OrderCount'].std().reindex(range(1,13))
months = range(1,13)
plt.plot(months, avg_orders, marker='o', color='tomato')
plt.fill_between(months, avg_orders - std_orders, avg_orders + std_orders, color='tomato', alpha=0.2)
plt.xticks(months, pd.to_datetime(pd.Series(months).astype(str), format='%m').dt.month_name().str[:3])
plt.title('Average Monthly Orders Across Years')
plt.xlabel('Month')
plt.ylabel('Average Order Count')

# Annotate peak month
peak_month = int(avg_orders.idxmax())
peak_val = avg_orders.max()
plt.annotate(f'{pd.to_datetime(str(peak_month), format="%m").month_name()} ({int(peak_val)})',
             xy=(peak_month, peak_val), xytext=(peak_month, peak_val + std_orders.max()*1.1),
             ha='center', arrowprops=dict(arrowstyle='->'))

plt.tight_layout()
plt.show()
```

**INSIGHTS** : There is an seasonal trends in order volume with August 1733 having the average monthly orders across Years.


# 5. How does total sales vary over time?

```python
# Visualize how total sales vary over time (monthly totals with rolling averages)
rev = revenue_by_month.copy()
rev['MonthYear'] = pd.to_datetime(rev['MonthYear'])
rev = rev.sort_values('MonthYear').set_index('MonthYear')

# Rolling averages
rev['rolling_3m'] = rev['TotalAmount'].rolling(3, min_periods=1).mean()
rev['rolling_12m'] = rev['TotalAmount'].rolling(12, min_periods=1).mean()

# Plot
import matplotlib.ticker as mtick
plt.figure(figsize=(14,6))
plt.plot(rev.index, rev['TotalAmount'], label='Monthly Total Revenue', marker='o', color='tab:blue')
plt.plot(rev.index, rev['rolling_3m'], label='3-month Rolling Mean', linestyle='--', color='orange')
plt.plot(rev.index, rev['rolling_12m'], label='12-month Rolling Mean', linestyle=':', color='green')

# Highlight peak month
peak_date = rev['TotalAmount'].idxmax()
peak_val = rev['TotalAmount'].max()
plt.scatter([peak_date], [peak_val], color='red', zorder=5, s=80)
plt.annotate(f'Peak: {peak_date.strftime("%Y-%m")}\n${peak_val:,.0f}',
             xy=(peak_date, peak_val), xytext=(0, 40), textcoords='offset points',
             ha='center', arrowprops=dict(arrowstyle='->', color='red'))

# Formatting
plt.title('Total Sales Over Time (Monthly Revenue)')
plt.xlabel('Month-Year')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

**INSIGHTS** : The total sales vary over time having the peak at May 2024 at a whopping One Million six hundred forty two thousand six hundred ten ($1,642,610)
